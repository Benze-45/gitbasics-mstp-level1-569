### Representing a sine wave
      * A=sine(wt)
             A=Amplitude
             w=frequency 
             t=time

```python
import numpy as np
import matplotlib.pyplot as pl
Time=np.arange(1,10,0.1)
Amplitude=np.sin(Time)
pl.grid()
pl.plot(Time,Amplitude)
pl.title('Sine Wave')
pl.xlabel('Time')
pl.ylabel('Amplitude')
pl.show()
```
